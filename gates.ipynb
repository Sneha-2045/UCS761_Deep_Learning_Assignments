```python
import numpy as np
```

```python
def step_activation(x):
    if x >= 0:
        return 1
    else:
        return 0
```

```python
def train_perceptron(X, Y, lr=0.1, epochs=100):

    weights = np.zeros(X.shape[1])
    bias = 0

    for epoch in range(epochs):

        for i in range(X.shape[0]):
            z = np.dot(weights, X[i]) + bias

            y_pred = step_activation(z)

            error = Y[i] - y_pred

            weights = weights + lr * error * X[i]
            bias = bias + lr * error

    return weights, bias
```

```python
def predict(X, weights, bias):

    predictions = []

    for i in range(X.shape[0]):

        z = np.dot(weights, X[i]) + bias
        y_pred = step_activation(z)

        predictions.append(y_pred)

    return np.array(predictions)
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
```

AND GATE

```python
Y_and = np.array([0,0,0,1])
```

```python
weights, bias = train_perceptron(X, Y_and)
```

```python
pred = predict(X, weights, bias)
```

```python
print("AND Gate")
print("Predictions:", pred)
```

OR Gate

```python
Y_or = np.array([0,1,1,1])
weights, bias = train_perceptron(X, Y_or)
pred = predict(X, weights, bias)
print("\nOR Gate")
print("Predictions:", pred)
```

NAND Gate

```python
Y_nand = np.array([1,1,1,0])
weights, bias = train_perceptron(X, Y_nand)
pred = predict(X, weights, bias)
print("\nNAND Gate")
print("Predictions:", pred)
```

NOR gate

```python
Y_nor = np.array([1,0,0,0])

weights, bias = train_perceptron(X, Y_nor)

pred = predict(X, weights, bias)

print("\nNOR Gate")
print("Predictions:", pred)
```

XOR gate

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

Y = np.array([[0],[1],[1],[0]])
```

```python

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def sigmoid_derivative(x):
    return x * (1 - x)

```

```python
lr = 0.3
epochs = 20000
```

```python
W1 = np.random.randn(2,2)
b1 = np.zeros((1,2))

W2 = np.random.randn(2,1)
b2 = np.zeros((1,1))
```

```python
for epoch in range(epochs):

    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, W2) + b2
    y_pred = sigmoid(z2)

    error = Y - y_pred

    d_output = error * sigmoid_derivative(y_pred)

    d_hidden = np.dot(d_output, W2.T) * sigmoid_derivative(a1)

    W2 += lr * np.dot(a1.T, d_output)
    b2 += lr * np.sum(d_output, axis=0, keepdims=True)

    W1 += lr * np.dot(X.T, d_hidden)
    b1 += lr * np.sum(d_hidden, axis=0, keepdims=True)
```

```python
print("XOR Results")
for x in X:

    h = sigmoid(np.dot(x, W1) + b1)
    o = sigmoid(np.dot(h, W2) + b2)

    y = 1 if o >= 0.5 else 0

    print(x[0], x[1], "->", y)
```
